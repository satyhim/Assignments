```python
import pandas as pd
import numpy as np
import datetime
from datetime import date 
data=pd.read_csv("/content/drive/MyDrive/Python/Assignement_cutoff3.csv")
```

```python
data
```

```python
df1=data.groupby(['customer ID', 'Date']).agg(max=('Brand Count', 'max'),total=('Brand Count', 'sum')).reset_index()
df1
```

```python
df1['Percentage cutoff']= (df1['max']/df1['total'])*100
df1
```

```python
df2=data.groupby(['customer ID','Date']).agg(min=('Brand Count', 'min'),total=('Brand Count', 'sum')).reset_index()
df2
```

```python
df2['Percentage cutoff']= (df2['min']/df2['total'])*100
df2
```

```python
Final = pd.concat([df1 , df2], axis=1, sort=False)
Final
print(df1, "\n\n\n", df2)
```
